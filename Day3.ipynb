### 6. 데이터 표 병합하기
두 개의 엑셀파일에서 데이터를 불러온 뒤, 데이터 병합

### pd.merge()
pd.merge(A, B, how = 'left', A의 테이블 기준값 지정, B의 테이블 기준값 지정)
- ex)pd.merge(A, B, how = 'right', left_on = "A컬럼명", right_index = True)
    - right  :A에 속한 데이터 기준
    - left : B에 속한 데이터 기준
    - inner : 양쪽에 모두 속한 데이터 기준
    - outter : 모든 데이터 기준    
    

- left_on : A 병합 기준 컬럼 지정
- left_index = True : A 병합기준 인덱스로 지정
- right_on : B 병합기준 컬럼 지정
- right_index = True : B 병합기준 인덱스로 지정

```python
fpath = './data/exam.xlsx'
A = pd.read_excel(fpath, index_col = '번호')
A
```

```python
fpath = './data/exam_extra.xlsx'
B = pd.read_excel(fpath, index_col = '번호')
B
```

```python
# A테이블의 키값을 기준(how = 'left')
# A테이블은 인덱스를 키 값으로, B테이블은 인덱스를 키 값으로 병합한다.
#>> A테이블의 인덱스를 키값으로 하므로 B테이블의 4,5번은 생략된다.

total = pd.merge(A, B, how = 'left', left_index = True, right_index = True)
total
```

```python
#B테이블의 키 값을 기준으로
#A테이블은 인덱스를 키 값으로, B테이블은 인덱스를 키 값으로

total = pd.merge(A, B, how = 'right', left_index = True, right_index = True)
total
```

```python
#A,B 테이블을 A,B 테이블 양쪽에 모두 존재하는 키 값을 기준
# A 테이블은 인덱스를 키 값으로, B테이블은 인덱스를 키 값으로

total = pd.merge(A, B, how = 'inner', left_index = True, right_index = True)
total
```

```python
#A,B 테이블 양쪽에 한번이라도 존재하는 모든 키 값을 기준으로
#A 테이블, B 테이블 모두 인덱스를 키 값으로 병합

total = pd.merge(A, B, how = 'outer', left_index = True, right_index = True)
total
```
